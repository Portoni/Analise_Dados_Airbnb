#### O que é o Airbnb
Airbnb é um serviço que permite proprietários de imóveis a alugar suas casas, apartamentos ou quartos para viajantes.
Este serviço se tornou muito popular pois não são todas as pessoas que podem pagar pela estadia em um hotel, ou até mesmo achar algum quarto vago nos hoteis das cidades mais turísticas. É por isso que o Airbnb cresceu muito desde sua criação em 2018, está presente em mais de 190 países e mais de 4 mil cidades, mais de 500 milhões de pessoas já alugaram alguma propriedade utilizando o Airbnb.

#### Como funciona - hóspede
Para alugar um Airbnb, o hóspede deve criar uma conta no applicativo, dentro do app, o usuário irá fornecer suas credenciais. Existe uma listagem de imóveis que podem ser filtrados por diversas categorias, preço, nota das críticas, tempo da estadia, quantidade de pessoas e tipo de imóvel. Você pode reservar o local e só paga depois que o proprietário aceitar sua estadia.

#### Como funciona - proprietário
Para listar seu imóvel no aplicativo é totalmente gratuito. Você deve fazer uma descrição da sua propriedade que será alugada, para melhorar as chances de compatibilidade com os usuários certos. O preço é estabelecido pelo proprietário e ele que aceita ou não a estadia das pessoas no imóvel. Porém, existe a possibilidade de reserva automática, caso o hóspede passe nos critérios listados pelo anfitrião. Além disso, o Airbnb cobre contra danos causados por clientes.

## Limpeza de Dados

Antes de analisar mais a fundo o dataset, vamos realizar uma das etapas mais importantes, a limpeza de dados.

### Dados Numéricos
Vamos criar um dataset somente com os dados númericos e verificar os dados nulos.

```python
num_listings = listings.select_dtypes(include=[np.number]).copy()
num_listings.isnull().sum().sort_values(ascending=False)
```

A coluna `neighbourhood_group` é composta praticamente por números nulos, então podemos eliminá-la. Já a coluna `reviews_per_month`, as colunas nulas são devido a imóveis que ainda não receberam nenhuma nota, então podemos preencher esses dados com `0`.

```python
num_listings.drop('neighbourhood_group', axis=1, inplace=True)
num_listings['reviews_per_month'].fillna(0, inplace=True)
```

As colunas `id` e `host_id` contém muito valores únicos e servem somente para diferenciar os imóveis ou os proprietários, portanto iremos deletá-las. Também a coluna `number_of_reviews_ltm`, não tem propósito algum para nossa análise, além de não sabermos o real significado dos valores presentes nela.

```python
columns = ['id', 'host_id', 'number_of_reviews_ltm']
num_listings.drop(columns, axis=1, inplace=True)
```

### Dados não numéricos

Agora vamos limpar as colunas não numéricas

```python
non_numeric = listings.select_dtypes(exclude=[np.number]).copy()
non_numeric.isnull().sum()
```

As colunas `name` e `host_name` não são importantes para nossa análise, e `license` é predominada por valores nulos, então iremos remover estas features.

```python
columns = ['name', 'host_name', 'license', 'last_review']
non_numeric.drop(columns, axis=1, inplace=True)
```

Com os dados limpos, vamos unificar o dataset novamente.

```python
list_clean = num_listings.merge(non_numeric, left_index=True, right_index=True)
```

Agora que juntamos novamente os dois datasets, é importante avaliar a integridade de alguns valores, como por exemplo os preços da coluna `price`. Não faz sentido alguém alugar um imóvel de graça, ou seja, todos os preços devem ser maior que `0`. Então vamos eliminar as linhas que possuem valor igual a `0`.

```python
drop_index = list_clean[list_clean['price'] == 0].index
list_clean.drop(drop_index, inplace=True)
```

## Detectando Outliers

Outliers são dados muito discrepantes que podem causar resultados errôneos em análises estatísticas, por isso é muito importante tratá-los.

### Distribuição dos valores

Vamos primeiro visualizar a distribuição dos valores das variáveis numéricas.

```python
import matplotlib.pyplot as plt
%matplotlib inline

def hist_plot(df, columns):
    for col in columns:
        df.hist(col)
        plt.show()
        
columns = ['price', 'minimum_nights']
hist_plot(list_clean, columns)
```

Essas colunas possuem uma distribuição assimétrica para direita severa, fortes indícios de um outliers, vamos analisar mais afundo.

### Boxplot

Vamos analisar os outliers por meio dos gráficos de caixa.

```python
def box_plot(df, columns):
    for col in columns:
        plt.boxplot(df[col], vert=False)
        plt.title(col)
        plt.show()

box_plot(list_clean, columns)
```

Existem muitos outliers nesse dataset, mas como eles representam uma pequena porcentagem dos nossos dados, podemos eliminar esses pontos sem alterar profundamente nossa análise.

### Removendo Outlier com método IQR
Existem diversas formas de eliminar os outliers de um conjunto de dados, neste projeto iremos utilizar o método Interquartile Range (IQR).

O IQR é a diferença da distribuição dos valores da variável entre o terceiro quartil e o primeiro quartil. Esse método define um limite superior e um limite inferior, qualquer valor que esteja 1,5 vezes acima do limite superior ou 1,5 vezes abaixo do limite inferior, é considerado um outlier.

```python
list_iqr = list_clean.copy()

def iqr(df, columns, scale=1.5):
    """Define os limites inferior e superior do método IQR
    """
    for col in columns:
        q1 = df[col].quantile(.25)
        q3 = df[col].quantile(.75)
        iqr = q3 - q1
        sup_limit = q3 + scale * iqr
        inf_limit = q1 - scale * iqr
        df.drop(df[df[col] > sup_limit].index, inplace=True)
        df.drop(df[df[col] < inf_limit].index, inplace=True)
    return df

list_iqr = iqr(list_iqr, columns)
```

Vamos ver novamente como ficou nossa distribuição de valores depois da remoção de outliers.

```python
hist_plot(list_iqr, columns)
```

```python
box_plot(list_iqr, columns)
```
